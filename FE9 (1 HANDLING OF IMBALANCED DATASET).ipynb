```python
import pandas as pd
import numpy as np
```

```python
Data=pd.read_csv("Credit_Card.csv")
```

```python
Data
```

```python
Data.isnull().sum()
```

```python
Data['Class'].value_counts()
```

```python
X=Data.drop('Class',axis=1)
y=Data['Class']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# 1.CROSS VALIDATION AND KFOLD
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
```

```python
LR=LogisticRegression()
Grid={'C':10.0 **np.arange(-2,3),'penalty':['l1','l2']}
CV=KFold(n_splits=5,random_state=None,shuffle=False)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7)
```

```python
CLF=GridSearchCV(LR,Grid,cv=CV,n_jobs=-1,scoring='f1_macro')
CLF.fit(X_train,y_train)
```

```python
y_pred=CLF.predict(X_test)
y_pred
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# 2.CLASS WEIGHTS IN ENSEMBLE
```

```python
Data=pd.read_csv('Credit_Card.csv',nrows=10000)
```

```python
Data
```

```python
Data['Class'].value_counts()
```

```python
# EXAMPLE: 0--->10000 1---->100 SO, 1:100 (GIVE 100 TIMES LESS IMPORTANT TO ZERO ONES)
# Class_Weight=dict({0:1,1:100})
from sklearn.ensemble import RandomForestClassifier
Classifier=RandomForestClassifier(class_weight="balanced")
```

```python
X=Data.drop('Class',axis=1)
y=Data['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7)
```

```python
Classifier.fit(X_train,y_train)
```

```python
y_pred=Classifier.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
