```python
# GAUSSIAN TRANSFORMATION
# 1.lOGARITHIMIC
# 2.RECIPROCAL
# 3.SQUARE ROOT
# 4.EXPONENTIAL
# 5.BOXCOX
```

```python
Data=pd.read_csv('Titanic.csv',usecols=['Age','Fare','Survived'])
```

```python
Data['Age'].fillna(Data['Age'].median(),inplace=True)
```

```python
Data.isnull().sum()
```

```python
# CHECK WHETHER THE DISTRIBUTION IS NORMAL OR GAUSSIAN (Q-Q PLOT) 
```

```python
import scipy.stats as stat
import pylab
```

```python
def Plot_Data(Data,Feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    Data[Feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(Data[Feature],dist='norm',plot=pylab)
    plt.show()
```

```python
Plot_Data(Data,'Age')
```

```python
# 1.LOGARITHIMIC TRANSFORMATION : WORKS BEST ON SKEWED PLOT(LEFT OR RIGHT)
```

```python
import numpy as np
```

```python
Data['Age_Log']=np.log(Data['Age'])
Plot_Data(Data,'Age_Log')
```

```python
# 2.RECIPROCAL TRANSFORMATION
```

```python
Data['Age_Reciprocal']=1/(Data['Age'])
Plot_Data(Data,'Age_Reciprocal')
```

```python
# 3.SQUAREROOT TRANSFORMATION
```

```python
Data['Age_Square']=Data['Age']**(1/2)
Plot_Data(Data,'Age_Square')
```

```python
# 4.EXPONENTIAL TRANSFORMATION
```

```python
Data['Age_Exponential']=Data['Age']**(1/1.2)
Plot_Data(Data,'Age_Exponential')
```

```python
# 5.BOXCOX TRANSFORMATION
```

```python
Data['Age_Boxcox'],Parameters=stat.boxcox(Data['Age'])
```

```python
Parameters
```

```python
Plot_Data(Data,'Age_Boxcox')
```

```python
Plot_Data(Data,'Fare')
```

```python
Data['Fare_Log']=np.log1p(Data['Fare'])
Plot_Data(Data,'Fare_Log')
```

```python
Data['Fare_Boxcox'],Parameters=stat.boxcox(Data['Fare']+1)
Plot_Data(Data,'Fare_Boxcox')
```
